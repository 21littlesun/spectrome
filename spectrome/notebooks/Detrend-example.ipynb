```python
from scipy import signal
import numpy as np
randgen = np.random.RandomState(9)
npoints = 1000
noise = randgen.randn(npoints)
x = 3 + 2*np.linspace(0, 1, npoints) + noise
(signal.detrend(x) - noise).max() < 0.01
True
```
